```python
import pandas as pd
import numpy as np
```

```python
gameid = '0021500002'
events = pd.read_csv('../../Volumes/KevSSDrive/data/events/{}.csv'.format(gameid))
all_movements = pd.read_csv('../../Volumes/KevSSDrive/NBA/data/csv/{}.csv'.format(gameid))
events = events.drop(labels=0,axis = 'index')
events = events.rename(columns = {'EVENTNUM': 'event_id'})
events = events[['event_id', 'EVENTMSGTYPE', 'EVENTMSGACTIONTYPE', 'SCORE']]
total_motion = all_movements.merge(events, how='left', on='event_id')
total_motion.to_csv('merged_{}.csv'.format(gameid), index=False)
```
